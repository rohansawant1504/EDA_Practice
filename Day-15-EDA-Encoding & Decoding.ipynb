## By using get_dummies

```python
df1 = pd.get_dummies(df.area_type)
```

```python
df1
```

### now we get encoding value for area , so we can do concat this df1 in our final data, so we can drop area_type 

```python
final_df = pd.concat([df,df1],axis=1).drop('area_type',axis=1)                   
```

```python
final_df
```

### now in size feature already encode 2bhk,3bhk etc. but we required only numrical value 

```python
final_df['size'] = final_df['size'].apply(lambda x:x.split(' ')[0])  ## here erroe occures because our size in dtype object 
```

```python
final_df['size'] = final_df['size'].apply(lambda x:str(x).split(' ')[0]) ### here only take this size in str 
```
