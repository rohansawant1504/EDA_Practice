## Pie chart

```python
team=['csk','kkr','mi','ds','rr','rcb']
score=[260,180,135,156,234,180]
```

```python
plt.pie(score,labels=team,shadow=True)
```

```python
plt.pie(score,labels=team,shadow=True,explode=(0.30,0,0,0,0))
```

```python
explode=(0,0,0.3,0)
```

```python
# Creating the pie chart
plt.pie(
    team,
    labels=score,  # Provide meaningful labels
    shadow=True,
    explode=explode, # Length of explode must match length of `score`
    autopct='%1.1f%%' # Add percentages to the slices
)
plt.title('Score Distribution')  # Add a title
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
score = [25, 35, 20, 10, 10]  # Values for the pie chart
labels = ['A', 'B', 'C', 'D', 'E']  # Labels for the slices

# Correct explode tuple: one value for each slice
explode = (0.10, 0, 0, 0, 0)

# Creating the pie chart
plt.pie(
    score,
    labels=labels,  # Provide meaningful labels
    shadow=True,
    explode=explode,  # Correct length
    autopct='%1.1f%%'  # Add percentages to the slices
)
plt.title('Score Distribution')  # Add a title
plt.show()
```
