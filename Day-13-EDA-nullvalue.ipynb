### now to check wether outlier present or not in data set for bath

```python
# value count for outlier detection
# min value = Q1-[1.5*IQR]
# max value = Q3+[1.5*IQR]
# IQR = Q3-Q1
```

```python
df.describe() # here outlier present so we can take median value for above farmula
```

```python
min_value = m-1.34*3
```

```python
min_value     # below this outlier
```

```python
max_value = m+1.34*3
```

```python
max_value    # above this outlier
```

```python
df['bath'].median()
```

```python
df['bath'] = df['bath'].fillna(df['bath'].mean())  # fill null value by median value 
```

```python
df.isnull().sum()
```

```python
df['balcony'] = df['balcony'].fillna(df['balcony'].median())  # fill null value by median value 
```
