###  Here we stored independent feature in X and dependent feature in y fro training and testing data

## In real time if we recieved data then we also recieved Metadata sheet & in that sheet mention which dependent and independent data.

### now we will do  X , Y balance  data

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
obj = RandomOverSampler(sampling_strategy = 0.75)
```

```python
X_new,y_new = obj.fit_resample(X,y)
```

```python
y_new.value_counts()
```

#### now we can do downsampling

```python
from imblearn.under_sampling import NearMiss
```

```python
obj = NearMiss(sampling_strategy = 0.75)
```

```python
X_under,y_under = obj.fit_resample(X,y)
```

```python
y_under.value_counts()
```
